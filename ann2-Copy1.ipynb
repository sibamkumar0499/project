```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sbn
```

```python
path = "normal.csv"
data = pd.read_csv(path,  low_memory=False) 
```

```python
data.head()
```

```python
data.describe()
```

```python
train_x = data.iloc[:,1:]
train_x.head()
```

```python
train_y = data['label']
train_y.head()
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LeakyReLU, PReLU, ELU
from keras.layers import Dropout
```

```python
model = Sequential([
    Dense(units = 1024, kernel_initializer = 'he_normal', input_shape=(99,), activation = 'relu'),
#     Dropout(0.5),
    Dense(units = 512, kernel_initializer = 'he_normal',  activation = 'relu'),
#     Dropout(0.5),
    Dense(units = 512, kernel_initializer = 'he_normal', activation = 'relu'),
#     Dropout(0.5),
    Dense(units = 256, kernel_initializer = 'he_normal', activation = 'relu'),
#     Dropout(0.5),
#     Dense(units = 71, kernel_initializer = 'glorot_uniform',  activation = 'sigmoid')
    Dense(units = 71, kernel_initializer = 'glorot_uniform',  activation = 'softmax')
])
#he_uniform
model.compile(optimizer = 'adam', 
              loss= 'sparse_categorical_crossentropy',
              metrics=['accuracy'])

# model.fit(x_train, y_train, validation_split=0.30, batch_size=100, epochs=20)
hist = model.fit(train_x, train_y, validation_data=(train_x.to_numpy(), train_y.to_numpy()), batch_size=512, epochs=100)
hist
```

```python
history = hist
print(history.history.keys())
# summarize history for accuracy
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['training accuracy', 'validation accuracy'], loc='upper left')
plt.show()
# summarize history for loss
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['training loss', 'validation loss'], loc='upper left')
plt.show()
```

```python
path = "bag.csv"
data_bg = pd.read_csv(path, low_memory=False)
```

```python
bg_test_x = data_bg.iloc[:,1:]
bg_test_x.head()
```

```python
bg_test_y = data_bg['label']
bg_test_y.head()
```

```python
y_pred = model.predict(bg_test_x)
```

```python
y_pred[0]
```

```python
np.argmax(y_pred[0])
```

```python
output_labels = [np.argmax(i) for i in y_pred]
output_labels
```

```python
import sklearn
from sklearn.metrics import accuracy_score
```

```python
accuracy = accuracy_score(bg_test_y, output_labels, normalize=True, sample_weight=None)
accuracy
```
